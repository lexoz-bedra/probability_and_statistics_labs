–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –≤—ã–±–æ—Ä–∫—É, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ $n$ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö —Å–ª—É—á–∞–π–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω, —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö –ø–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–º—É –∑–∞–∫–æ–Ω—É —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ $-Œ∏$ –∏ $Œ∏$. –ü–æ –Ω–µ–π –æ—Ü–µ–Ω–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä $\theta^2$.


$$X_1, \cdots, X_n \sim U[-\theta, \theta]$$


–í—ã—á–∏—Å–ª–∏–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é:
\
$$EX_i = \int_{-‚àû}^\infty x‚ãÖp(x)dx = \int_{-\theta}^\theta x‚ãÖ\frac{1}{2\theta}dx = \frac{x^2}{4\theta}\Biggr|_{-\theta}^\theta = \frac{\theta^2-(-\theta)^2}{4\theta} = 0$$

\

$$EX_{i}^2 = \int_{-\infty}^\infty x^2‚ãÖp(x)dx = \int_{-\theta}^\theta x^2‚ãÖ\frac{1}{2\theta}dx = \frac{x^3}{6\theta}\Biggr|_{-\theta}^\theta = \frac{2\theta^3}{6\theta} = \frac{\theta^2}{3}$$

\

$$Var(X_i) = EX_{i}^2 - (EX_{i})^2 = \frac{\theta^2}{3} - 0 = \frac{\theta^2}{3}$$

\

–ü—Ä–∏—Ä–∞–≤–Ω—è–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –º–æ–º–µ–Ω—Ç—ã –∫ –≤—ã–±–æ—Ä–æ—á–Ω—ã–º:

\

\begin{cases}
\hat{\alpha_1} = 0
\\\hat{\beta_2} = \frac{\hat{\theta}^2}{3}
\end{cases}

\

\begin{cases}
\overline{X_{n}} = 0
\\S_{*}^2 = \frac{\hat{\theta}^2}{3}
\end{cases}

\
$$\hat{\theta}^2 = 3S_{*}^2$$

–ú—ã –ø–æ–ª—É—á–∏–ª–∏ –æ—Ü–µ–Ω–∫—É –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\theta^2$.

–ù–∞–π–¥—ë–º —Å–º–µ—â–µ–Ω–∏–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –æ—à–∏–±–∫—É (MSE) –¥–ª—è –æ—Ü–µ–Ω–∫–∏ $\hat{\theta}^2$.



* **–°–º–µ—â–µ–Ω–∏–µ**

$$b_n(Œ∏^2) = E(\hat{\theta}^2) - \theta^2$$
\
$$E(\hat{\theta}^2) = E(3S_{*}^2) = 3E(S_{*}^2) = 3E(\hat{\alpha_2} - \hat{\alpha_1}^2) = 3E(\hat{\alpha_2}) - 3E(\hat{\alpha_1}^2) = 3E(\overline{X_{n}^2}) - 3E(\overline{X_{n}}^2) = 3E(X_{i}^2) - 0 = 3 \cdot \frac{\theta^2}{3} = \theta^2$$

\

$$b_n(\theta^2) = \theta^2 - \theta^2 = 0$$

* **–î–∏—Å–ø–µ—Ä—Å–∏—è**
\
$$Var(\hat{\theta}^2) = E(\hat{\theta}^4) - E(\hat{\theta}^2)^2$$
\
$$E(\hat{\theta}^4) = E(9S_{*}^4) = 9E(\hat{\alpha_4} - 4\hat{\alpha_1}\hat{\alpha_3} + 6\hat{\alpha_2}\hat{\alpha_1}^2 - 3\hat{\alpha_1}^4) = 9E(\hat{\alpha_4}) - 36E(\hat{\alpha_1}\hat{\alpha_3}) +54E(\hat{\alpha_2}\hat{\alpha_1}^2) - 27E(\hat{\alpha_1}^4) = 9E(\overline{X_{n}^4}) - 36E(\overline{X_n}\overline{X_{n}^3}) +54E(\overline{X_{n}^2}\overline{X_n}^2) - 27E(\overline{X_{n}}^4)$$
\
$$E(\overline{X_n}) = E(X_i) = 0$$
\
$$E(\overline{X_{n}^4}) = E(X_{i}^4) = \int_{-\infty}^\infty x^2‚ãÖp(x)dx = \int_{-\theta}^\theta x^4‚ãÖ\frac{1}{2\theta}dx = \frac{x^5}{10\theta}\Biggr|_{-\theta}^\theta = \frac{2\theta^5}{10\theta} = \frac{\theta^4}{5}$$

\
$$E(\hat{\theta}^4) = 9E(\overline{X_{n}^4}) - 0 + 0 - 0 = 9 \cdot \frac{\theta^4}{5} = \frac{9\theta^4}{5}$$

\
$$Var(\hat{\theta}^2) = \frac{9\theta^4}{5} - \theta^4 = \frac{4\theta^4}{5}$$
* **MSE**

$$MSE(\hat{\theta}^2) = E((\hat{\theta}^2‚àí\theta^2)^2) = Var(\hat{\theta}^2) + b_{n}(\hat{\theta}^2)^2 = \frac{4\theta^4}{5}$$

## –°–≤–æ–π—Å—Ç–≤–∞ –æ—Ü–µ–Ω–∫–∏

* **–ù–µ—Å–º–µ—â—ë–Ω–Ω–æ—Å—Ç—å**
$$E(\hat{\theta^2}) = \theta^2$$

–í –ø—Ä–µ–¥—ã–¥—É—â–µ–º –ø—É–Ω–∫—Ç–µ –º—ã –≤—ã—á–∏—Å–ª–∏–ª–∏ —Å–º–µ—â–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –æ–∫–∞–∑–∞–ª–æ—Å—å —Ä–∞–≤–Ω–æ –Ω—É–ª—é (—Ç–æ –µ—Å—Ç—å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å –∏—Å—Ç–∏–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞)

* **–≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å**
$$MSE(\hat{\theta}) = min(MSE(\hat{\theta}_i))$$

–û—Ü–µ–Ω–∫–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞, –∫–æ–≥–¥–∞ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–∞—è MSE, —Ç–æ –µ—Å—Ç—å –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è, –∏ —Å–º–µ—â–µ–Ω–∏–µ –∫–∞–∫ –º–æ–∂–Ω–æ –º–µ–Ω—å—à–µ. –ù–∞—à–∞ –æ—Ü–µ–Ω–∫–∞ –Ω–µ—Å–º–µ—â—ë–Ω–Ω–∞—è, –ø–æ—ç—Ç–æ–º—É –æ–Ω–∞ –±—É–¥–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞, –µ—Å–ª–∏ –µ—ë –¥–∏—Å–ø–µ—Ä—Å–∏—è –Ω–∞–∏–º–µ–Ω—å—à–∞—è –∏–∑ –≤–æ–∑–º–æ–∂–Ω—ã—Ö. –ü—Ä–æ–≤–µ—Ä–∏–º —ç—Ç–æ —Å –ø–æ–º–æ—â—å—é –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –†–∞–æ-–ö—Ä–∞–º–µ—Ä–∞:

$$Var(\hat{\theta}^2) \ge \frac{1}{I(\theta^2)}$$
\
$$I(\theta^2) = E_{\theta^2}(\frac{\partial{lnL(\theta^2, x)}}{\partial{\theta^2}})^2 = E_{\theta^2}(\frac{\partial{ln\frac{1}{2\theta}}}{\partial\theta^2})^2 = E_{\theta^2}(-\frac{1}{2\theta^2})^2 = \frac{1}{4\theta^4}$$
\
$$Var(\theta^2) \ge 4\theta^4$$

–ß—Ç–æ-—Ç–æ –Ω–µ —Å—Ö–æ–¥–∏—Ç—Å—è..........

–û—Ü–µ–Ω–∫–∞ –Ω–∞—Å—Ç–æ–ª—å–∫–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞, —á—Ç–æ –ø–æ–±–µ–¥–∏–ª–∞ –¥–∞–∂–µ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –†–∞–æ-–ö—Ä–∞–º–µ—Ä–∞!!!


* **–°–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ—Å—Ç—å**
$$\lim_{n \to \infty}‚Ñô(|\hat{\theta}_{n}^2 - \theta_n^2| ‚â• Œµ) = 0$$
\
$$P(|\hat{\theta_{n}^2} - \theta_{n}^2 | < Œµ) \to 1$$

–≠—Ç–æ –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å –≤ –≤–∏–¥–µ:

$$\lim_{n \to \infty} \hat{\theta_n}^2 = \theta_{n}^2$$

\

$$\lim_{n \to \infty} \hat{\theta_n}^2 = \lim_{n \to \infty} 3S_*^2 = 3\lim_{n \to \infty} S_*^2 = 3 \lim_{n \to \infty} \overline{X_n^2} - 3 \lim_{n \to \infty} \overline{X_n}^2 = 3E(X_i^2) - 3E(X_i)^2 = 3 \cdot Var(X_i) = 3 \cdot \frac{(\theta - (-\theta))^2}{12} = \frac{12\theta^2}{12} = \theta_n^2$$

–û—Ü–µ–Ω–∫–∞ —Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–∞.



* **–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å**

$$\sqrt{n}(\hat{\theta}_n^2 - \theta_n^2) \to ùí©(0, œÉ^2)\  –ø—Ä–∏\  n \to \infty$$

–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫—É—é –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ–≤–µ—Ä–∏–º –≤ —Ö–æ–¥–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞.

## –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç

–ü—Ä–æ–≤–µ—Ä–∏–º –Ω–∞—à–∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ —Ä–∞—Å—á—ë—Ç—ã –Ω–∞ –Ω–∞–±–æ—Ä–µ —Å–ª—É—á–∞–π–Ω–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

–î–ª—è —ç—Ç–æ–≥–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø–æ 100 –≤—ã–±–æ—Ä–æ–∫ 8 —Ä–∞–∑–Ω—ã—Ö –æ–±—ä—ë–º–æ–≤ (10, 50, 100, 250, 750, 1000, 5000, 10000), –Ω–∞ –∫–∞–∂–¥–æ–π –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —Å–¥–µ–ª–∞–µ–º –æ—Ü–µ–Ω–∫—É $\hat{\theta}$ –ø–æ –≤—ã–≤–µ–¥–µ–Ω–Ω–æ–π —Ä–∞–Ω–µ–µ —Ñ–æ—Ä–º—É–ª–µ –∏ —Å—Ä–∞–≤–Ω–∏–º –µ—ë —Å —Ä–µ–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞: $\theta = 10$.

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –ø–æ–ª—É—á–∏–≤—à–∏—Ö—Å—è –≤—ã–±–æ—Ä–æ–∫ –∏ –æ—Ü–µ–Ω–∏–º, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏—Ö —Ä–∞—Å—á—ë—Ç–æ–≤. –í —á–∞—Å—Ç–Ω–æ—Å—Ç–∏, –Ω–∞–π–¥—ë–º –≤—ã–±–æ—Ä–∫–∏, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –æ—Ü–µ–Ω–∫–∞ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–æ–≤–æ–ª—å–Ω–æ —Å–∏–ª—å–Ω–æ (–º—ã –ø—Ä–æ–≤–µ—Ä–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ—Ä–æ–≥–æ–≤ –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞–∑–ª–∏—á–∏—è).

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
```

```python
theta = 10
sample_sizes = np.array([10, 50, 100, 250, 750, 1000, 5000, 10000])
m = 100
thresholds = [0.05, 0.1, 0.25, 1]
```

```python
for threshold in thresholds:
    biased_counts = []

    for n in sample_sizes:
        count = 0
        for _ in range(m):
            sample = np.random.uniform(low=-theta, high=theta, size=n)
            theta_hat = np.sqrt(3 * np.mean(sample**2) - 3 * np.mean(sample)**2)
            if abs(theta_hat - theta) > threshold:
                count += 1
        biased_counts.append(count)

    plt.plot(sample_sizes, biased_counts, marker='o', label=f'Threshold = {threshold}')

plt.xlabel('–û–±—ä–µ–º –≤—ã–±–æ—Ä–∫–∏')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫ —Å —Å–∏–ª—å–Ω—ã–º –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º')
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–º–µ—â–µ–Ω–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫ –æ—Ç –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –ø–æ—Ä–æ–≥–æ–≤')
plt.grid(True)
plt.legend()
plt.show()
```

```python
absolute_diffs = np.zeros((len(sample_sizes), m))

for i, n in enumerate(sample_sizes):
    for j in range(m):
        sample = np.random.uniform(low=-theta, high=theta, size=n)
        theta_hat_squared = 3 * np.var(sample)
        absolute_diffs[i, j] = np.abs(theta_hat_squared - theta**2)

mean_absolute_diffs = np.mean(absolute_diffs, axis=1)

plt.plot(sample_sizes, mean_absolute_diffs, marker='o')
plt.xlabel('–û–±—ä–µ–º –≤—ã–±–æ—Ä–∫–∏')
plt.ylabel('|$\\hat{\\theta}^2 - \\theta^2$|')
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∞–±—Å–æ–ª—é—Ç–Ω–æ–π —Ä–∞–∑–Ω–∏—Ü—ã –æ—Ü–µ–Ω–∫–∏ $\\theta^2$ –æ—Ç –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏')
plt.grid(True)
plt.show()
```

```python
estimates = []

for n in sample_sizes:
    sample_estimates = []
    for _ in range(m):
        sample = np.random.uniform(low=-theta, high=theta, size=n)
        theta_hat = np.sqrt(3 * np.mean(sample**2) - 3 * np.mean(sample)**2)
        sample_estimates.append(theta_hat)
    estimates.append(sample_estimates)

plt.figure(figsize=(12, 8))

for i, n in enumerate(sample_sizes):
    plt.subplot(2, 4, i+1)
    plt.hist(estimates[i], bins=30, density=False, alpha=0.6, color='b')

    plt.title(f'n = {n}')
    plt.xlabel('–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞')
    plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
